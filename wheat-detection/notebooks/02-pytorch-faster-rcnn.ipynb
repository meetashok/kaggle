```python
import torchvision
import torch
```

```python
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained=True)
```

```python
num_images = 4
```

```python
images = torch.rand(num_images, 3, 600, 300)
boxes = torch.rand(num_images, 11, 4)

labels = torch.randint(1, 91, (4, 11))
```

```python
labels
```

```python
images = list(image for image in images)
```

```python
targets = []

for i in range(len(images)):
    targets += [{"boxes": boxes[i], "labels": labels[i]}]
```

```python
output = model(images, targets)
```
